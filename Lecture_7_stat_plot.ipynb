```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([5.9, 7.2, 7.3, 6.3, 8.1, 6.8, 7.0, 7.6, 6.8, 6.5, 7.0, 6.3, 7.9,
        9.0, 8.2, 8.7, 7.8, 9.7, 7.4, 7.7, 9.7, 7.8, 7.7, 11.6, 11.3, 11.8, 10.70])

plt.plot(data,'bv')
plt.show()
```

```python
data_asc = np.sort(data)

plt.plot(data_asc,'ro')
plt.show()
```

```python
import seaborn as sns
sns.histplot(data, color='skyblue')
```

```python
plt.boxplot(data,showmeans='True')
plt.show()
```

```python
import scipy.stats as stats
import statsmodels.graphics.gofplots as sm

sm.qqplot(data, line='q')
plt.title("Normal Q-Q Plot")
```
