```python
import numpy 
from scipy.stats import rv_discrete

x = [0,1,2]
p_x = [1/2, 1/3, 1/6]
X = rv_discrete(values=(x, p_x))
```

```python
print(X.pmf(x))
```

```python
print(X.cdf(x))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots() # creates a single figure and axes

ax.plot(x, X.pmf(x),'o')
ax.vlines(x, 0, X.pmf(x))
```

```python
fig, ax = plt.subplots()
ax.step(x,X.cdf(x))

```
